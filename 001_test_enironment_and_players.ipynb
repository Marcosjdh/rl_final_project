# Implementar una partida completa

Muestreando el esapcio de acción con env.action_space.sample (uniformemente)

Notar que el estado es (board, player)

```python
done = False
# Resetear el entorno
(board, player) = env.reset()
while not done: # Mientras que el episodio no termine, ejecutar
    action = env.action_space.sample() # Muestrear acciones uniformemente
    print(board)
    print(f'acion: {action}')
    
    print(f'player: {player}')
    (board, player), reward, done, _ = env.step(action) # Ejecutar un paso
    
```

```python
board
```

Notar que muestrar uniformemente es un problema por que el episodio termina enseguida por la alta probabilidad de muestrar una acción no válida

# Implementar sample_valid_actions que muestrea uniformemente las acciones válidas

```python
def sample_valid_actions(state, player):
    
    valid_actions = np.argwhere(env.get_valid((state, player)) == 1) 
    sampled_valid_action = valid_actions[np.random.randint(len(valid_actions))]
    return sampled_valid_action
```

```python
np.argwhere(env.get_valid((board, player)) == 1)
```

```python
# Verificar que funcione OK
sample_valid_actions(board, player)
```

### Correr ahora muestrando las acciones válidas y verificar que los epidosiod terminan en una situación donde ningún jugador puede mover mas

```python
done = False
(board, player) = env.reset()
while not done:
    action = sample_valid_actions(board, player)
    print(board)
    print(f'acion: {action}')
    
    print(f'player: {player}')
    (board, player), reward, done, _ = env.step(action)
```
